### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?q="

name = []
clouds = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

print('Beginning Data Retrieval \n-----------------------------')

count = 1
Set = 1

for city in cities:
    
    try:
        print('Processing Record {} of Set {} | {}'.format(count,Set,city))
        
        request = requests.get(url + "{}".format(city) + "&units=imperial"+"&APPID={}".format(api_key)).json()
        name.append(request['name'])
        clouds.append(request['clouds']['all'])
        country.append(request['sys']['country'])
        date.append(request['dt'])
        humidity.append(request['main']['humidity'])
        lat.append(request['coord']['lat'])
        lng.append(request['coord']['lon'])
        max_temp.append(request['main']['temp_max'])
        wind_speed.append(request['wind']['speed'])
        
        count += 1
        
    except Exception as e:
        print('City not found. Skipping...')
        
    if count == 51:
        count = 1
        Set += 1

print('----------------------------- \nData Retrieval Complete \n-----------------------------')
```

```python
weather_df = pd.DataFrame({"City": name, "Cloudiness": clouds, "Country":country, 
                           "Date": date, "Humidity": humidity, "Lat":lat, "Lng":lng, 
                           "Max Temp": max_temp, "Wind Speed":wind_speed})
```

#### Latitude vs. Temperature Plot

```python
import time
from datetime import date

date = date.today()
date = date.strftime("%m/%d/%y")

plt.scatter(x=weather_df['Lat'],y=weather_df['Max Temp'],edgecolors='black',alpha=.75)
plt.title('City Latitude vs Max Temperatuere ({})'.format(date))
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.grid()

plt.savefig('LatvTemp.png')
```
